```python
#Outliers it works with the numerical data
#circles represents the outliers
#orange line is the meadian
import matplotlib.pyplot as plt
plt.boxplot(data['age'])
plt.show()
```

```python
data=data[(data['age']<=75)&(data['age']>18)]
```

```python
#label Encoding
#converts the categorical value into the numerical values
#we can done through sklearn and its a preprocessing
#craete an object for the labelencider
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
#fit.transform is a method to tranforms the categorical value into numerical value

le=LabelEncoder()
data['workclass']=le.fit_transform(data['workclass'])
data['marital-status']=le.fit_transform(data['marital-status'])
data['occupation']=le.fit_transform(data['occupation'])
data['relationship']=le.fit_transform(data['relationship'])
data['race']=le.fit_transform(data['race'])
data['gender']=le.fit_transform(data['gender'])
data['native-country']=le.fit_transform(data['native-country'])
```

```python
data
```

```python
# imported the necessary libraries
#imported the necessary datasets
#visuaalised the data in the terms of rows and columns
#Evaluated if any null values are presented in the dataset
#Replaced the special characters with the values
#visualised the outliers using graph viualisation
#removed the outliers,not useful information
#we have deleted the redundant values
#we encoded the dataset from categorial value to numerical value
#For the better understanding of data that can feed to machine
```

```python
#after encoding technique we have to split the data as training and testing data
# x will store all the labels except income , it is the input variable
# y will store only the income column , it is the output variable
x=data.drop(columns=['income'])
y=data['income']
```

```python
x
```

```python
y
```

```python
#fnlwgt is in a large range so it fix in a fixed range of 0-1 using min-max scaling
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x=scaler.fit_transform(x)
x
```

```python
#train_test_split will divide the training and testing data
#x_train is i/p training data
#x_test i/p testing data
#y_train is o.p training data
#y_test is o/p testing data
#test_size will split 80% and 20%
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1,stratify=y)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#importing machine learning algorithm
#we divide data based on the neighbour points is Supervised learning model
from sklearn.neighbors import KNeighborsClassifier
KNN=KNeighborsClassifier()
KNN.fit(x_train,y_train)
predict=KNN.predict(x_test)
predict #predicted value
```

```python
#
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predict)
```

```python
#importing machine learning algorithm
#we divide data based on the neighbour points is Supervised learning model
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(x_train,y_train)
predicts=LR.predict(x_test)
predicts #predicted value
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predicts)
```

```python
#adam is optimiser take the best sol'n
#
from sklearn.neural_network import MLPClassifier
NN=MLPClassifier(solver='adam',hidden_layer_sizes=(10,10),random_state=1)
NN.fit(x_train,y_train)
predictss=NN.predict(x_test)
predictss #predicted value
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictss)
```
